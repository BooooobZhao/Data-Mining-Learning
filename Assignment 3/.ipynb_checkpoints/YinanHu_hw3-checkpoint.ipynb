## Part 1 - Visualizing the Face Images

### 1. Visualize randomly selected 16 faces in a 4Ã—4 grid

```python
fig = plt.figure(figsize=(8, 8))
columns = 4
rows = 4
IMAGE_FOLDER = 'face_data/'
index = 1
dirnames = random.sample(os.listdir(IMAGE_FOLDER), 16)
for dirname in dirnames:
    filename = random.choice(os.listdir(IMAGE_FOLDER + dirname))
    path = '%s/%s/%s' % (IMAGE_FOLDER, dirname, filename)
    with open(path, 'rb') as image:
        img = cv2.imdecode(np.frombuffer(image.read(), np.uint8), cv2.IMREAD_GRAYSCALE)
    fig.add_subplot(rows, columns, index)
    plt.imshow(img, cmap="gray")
    index = index + 1
plt.show()
```
